```python
import pandas as pd

df = pd.read_csv("/MovieData-2018-2020.csv")
df = df.dropna(subset=['budget'])

# Menghitung frekuensi setiap genre
genre_counts = df['genre'].value_counts()

# Mengidentifikasi genre dengan hanya satu kemunculan
single_occurrence_genres = genre_counts[genre_counts == 1].index

# Menghapus baris yang berisi genre dengan hanya satu kemunculan
df_filtered = df[~df['genre'].isin(single_occurrence_genres)]

# Stratified Random Sampling berdasarkan genre
sample, _ = train_test_split(
    df_filtered, # Gunakan df_filtered setelah menghapus genre tunggal
    test_size=0.8,      # mengambil 20% sebagai sampel
    stratify=df_filtered['genre'], # Stratify juga pada df_filtered
    random_state=62     # 3 digit belakang NIM = 062
)

# Mengambil 30 data teratas
sample30 = sample.head(30)

# Menampilkan 30 data teratas
print("\n=== HEAD(30) HASIL SAMPLING ===")
print(sample30)

# Informasi data hasil sampling
print("\n=== INFO HASIL SAMPLING ===")
sample30.info()

# Statistik deskriptif hasil sampling
print("\n=== DESCRIBE HASIL SAMPLING ===")
print(sample30.describe())
```

```python
mean_budget = sample30['budget'].mean()

print("Estimasi Mean Budget Populasi =", mean_budget)
```

```python
from scipy.stats import t
import numpy as np

n = len(sample30)
x_bar = sample30['budget'].mean()
s = sample30['budget'].std()

SE = s / np.sqrt(n)

t_value = t.ppf(0.975, df=n-1)

ME = t_value * SE

lower = x_bar - ME
upper = x_bar + ME

print("Confidence Interval 95%")
print(f"({lower:.2f}, {upper:.2f})")
```

```python
import scipy.stats as stats
import math

# Mengambil nilai langsung dari sample30
x_bar = sample30['budget'].mean()
s = sample30['budget'].std()
n = len(sample30)

confidence_level = 0.95

alpha = (1 - confidence_level)
df = n - 1

standard_error = s / math.sqrt(n)

critical_t = stats.t.ppf(
    1 - alpha/2,
    df
)

margin_of_error = critical_t * standard_error

lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

print("Critical t-value: {:.2f}".format(critical_t))
print("Margin of Error: {:.2f}".format(margin_of_error))
print("Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))

print(
    "The {:.1%} confidence interval for the population mean is:"
    .format(confidence_level)
)

print(
    "between {:.2f} and {:.2f}"
    .format(lower_bound, upper_bound)
)
```
